```python
import matplotlib.pyplot as plt
%matplotlib inline

x0 = [1, 2, 3, 4, 5, 6, 7]
y0 = [9, 8, 7, 6, 5, 4, 3]
y1 = [3, 4, 0, 9, 9, 1, 0]
```
